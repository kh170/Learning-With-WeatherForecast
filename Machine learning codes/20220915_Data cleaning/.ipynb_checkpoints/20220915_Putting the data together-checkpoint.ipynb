```python
def data_extract(year,month_list):
    # year: year in string format
    # month_list: list of months in string format
    data_appended = pd.DataFrame() #collects all data of desired year and months
    for month in month_list:
        x = pd.read_csv(data_path+"//"+year+month+".csv")
        data_appended = data_appended.append(x)
    return data_appended
    
    
    
    
def string2datetime(time_col):
    time_list = [str.split(list(time_col)[i])[0] for i in range(len(time_col))]
    return pd.to_datetime(time_list)
```

```python
month_list = ['01','02','03','04','05','06','07','08','09','10','11','12']
data_2018 = data_extract("2018",month_list[8::])
data_2019 = data_extract("2019",month_list)
data_2020 = data_extract("2020",month_list)
data_2021 = data_extract("2021",month_list)
data_2022 = data_extract("2022",month_list[0:9])
print("2018:",len(data_2018),"\n2019:",len(data_2019),"\n2020:",len(data_2020),"\n2021:",len(data_2021),"\n2022:",len(data_2022))
```

```python
# Putting the data together
data = pd.concat([data_2018,data_2019,data_2020,data_2021,data_2022])
pd.to_csv()
```
